**Modification de la Dataset**

```python
import re
def age_to_range(age_str):
    # Use regex to extract age range
    match = re.search(r'(\d+)-(\d+)', age_str)
    if match:
        age_min = int(match.group(1))
        age_max = int(match.group(2))
        return age_min, age_max
    else:
        return None, None  # Handle cases where regex doesn't match

dataset[['age_min', 'age_max']] = dataset['age'].apply(lambda x: pd.Series(age_to_range(x)))
```

```python
dataset.drop('age', axis=1, inplace=True)
```

```python
dataset[['tumor_size_min', 'tumor_size_max']] = dataset['tumor-size'].apply(lambda x: pd.Series(age_to_range(x)))
```

```python
dataset[['inv_nodes_min', 'inv_nodes_max']] = dataset['inv-nodes'].apply(lambda x: pd.Series(age_to_range(x)))
```

```python
dataset.drop(['tumor-size', 'inv-nodes'], axis=1, inplace=True)
```

```python
dataset['node-caps'] = dataset['node-caps'].map({'no': 0, 'yes': 1})
```

```python
dataset['breast'] = dataset['breast'].map({'left': 0, 'right': 1})
```

```python
dataset['breast-quad'] = dataset['breast-quad'].map({'left_low': 0, 'left_up':1, 'right_low': 2, 'right_up': 3, 'central': 4})
```

```python
dataset['irradiat'] = dataset['irradiat'].map({'no': 0, 'yes': 1})
```

```python
dataset['Class'] = dataset['Class'].map({'no-recurrence-events': 0, 'recurrence-events': 1})
```

```python
dataset['menopause'] = dataset['menopause'].map({'premeno': 0, 'ge40': 1, 'lt40': 2})
```

```python
dataset['menopause'] = dataset['menopause'].replace(2, 0)
```

```python
df=pd.DataFrame(dataset)
```

**Répartition des data en train and test**

```python
from sklearn.model_selection import train_test_split
```

```python
df['node-caps']=df['node-caps'].fillna(df['node-caps'].mode()[0])
df['breast-quad']=df['breast-quad'].fillna(df['breast-quad'].mode()[0])
```

```python
df_train, df_test = train_test_split(df,test_size=0.2)
```

```python
X_train=df_train.drop(columns=['Class']).values
Y_train=df_train['Class'].values
X_test=df_test.drop(columns=['Class']).values
Y_test=df_test['Class'].values
```
